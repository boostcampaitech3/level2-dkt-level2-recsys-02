```python
import pandas as pd
import os
```

```python
output_dir = './output/'
lgbm_preds = pd.read_csv(output_dir+'exp3.csv')
cb_preds = pd.read_csv(output_dir+'cat_exp1.csv')
xgb_preds = pd.read_csv(output_dir+'xgb_exp1.csv')
```

```python
lw = 0.7
cw = 0.1
xw = 0.2

preds = lgbm_preds['prediction']*lw + cb_preds['prediction']*cw +  xgb_preds['prediction']*xw
```

```python
write_path = os.path.join(output_dir, "712.csv")
with open(write_path, 'w', encoding='utf8') as w:   
    print("writing prediction : {}".format(write_path))
    w.write("id,prediction\n")
    for id, p in enumerate(preds):
        w.write('{},{}\n'.format(id,p))
```
